<h3>Coverting Datatype</h3>

<b>Transform data so we can work with it effectively</b>

Converting datatype of "started_at" & "ended_at" columns to "datetime64" type.

```python
# Converting datatype of "started_at" & "ended_at" columns to "datetime64" type.
df['started_at'] = df['started_at'].astype('datetime64')
df['ended_at'] = df['ended_at'].astype('datetime64')
```

Converting datatype of "ride_id" to "string".

```python
# Converting datatype "ride_id", "start_station_name" and "end_station_name" to "string"
df['ride_id'] = df['ride_id'].astype('str')
df['start_staion_name'] = df['start_station_name'].astype('str')
df['ens_station_name'] = df['end_station_name'].astype('str')
```

Trim all strings columns in the dataframe

```python
# trim all strings of a dataframe
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
```

Converting columns to category.

```python
# Converting columns to category for saving memory usage
cols_to_exclude = ['rideable_type','start_station_name','start_station_id','end_station_name','end_station_id']
for col in df.columns:
    if df[col].nunique() < 900 and col in cols_to_exclude:
        df[col] = df[col].astype('category')
```

Coverting datatype of "member_casual" to new column 'member' as "booleans" type and delete "member_casual" column.

```python
# Coverting datatype of "member_casual" to new column as "booleans"
# Delete "member_casual" columns that are not needed
df['member'] = np.where(df['member_casual'] == 'member', True, False)
df = df.drop(columns=['member_casual'])
```

Creating New Column "ride_length" and changing its datatype to "int32".

```python
# Creating New Column "ride_length" and changing its datatype to "int32"
# In this column, each row contains the difference between "starting time" and "ending time" columns in minutes.
df['ride_length'] = (df['ended_at'] - df['started_at'])/pd.Timedelta(minutes=1)
df['ride_length'] = df['ride_length'].astype('int32')
```

```python
btm = pd.DataFrame(df.groupby(['date','rideable_type'], as_index=False).agg({'ride_id' : 'count'}))
btm = btm.assign(
    day_of_week = btm['date'].dt.strftime("%A"),
    day_week_num = pd.to_datetime(btm['date']).dt.dayofweek)

rtc = btm.loc[(btm['rideable_type'] == 'classic_bike')]
rtc = rtc.drop(['rideable_type'], axis=1)
rtc.columns = ['Date','Classic Bike', 'Day','Day_num']
rtc = rtc.groupby(['Day_num','Day'], as_index=True).agg({'Classic Bike' : np.mean})
rtc['Classic Bike'] = rtc['Classic Bike'].astype(int)

rtd = btm.loc[(btm['rideable_type'] == 'docked_bike')]
rtd = rtd.drop(['rideable_type'], axis=1)
rtd.columns = ['Date','Docked Bike','Day','Day_num']
rtd = rtd.groupby(['Day_num','Day'], as_index=True).agg({'Docked Bike' : np.mean})
rtd['Docked Bike'] = rtd['Docked Bike'].astype(int)

rte = btm.loc[(btm['rideable_type'] == 'electric_bike')]
rte = rte.drop(['rideable_type'], axis=1)
rte.columns = ['Date','Electric Bike','Day','Day_num']
rte = rte.groupby(['Day_num','Day'], as_index=True).agg({'Electric Bike' : np.mean})
rte['Electric Bike'] = rte['Electric Bike'].astype(int)

a = rtc
a['Docked Bike'] = rtd
a['Electric Bike'] = rte
a = a.reset_index()
a = a.drop(['Day_num'], axis=1)
a
```

<h4>Examining Casual cycling routes</h4>

```python
member = df.loc[(df['member'] == True)] # member only list
casual = df.loc[(df['member'] == False)] # casual only list
```

```python
# top 10 start & end station rides for casual
cs = casual.groupby(['start_station_name']).agg(
    {"ride_length" : ["count","mean"],'distance':'mean','start_lat' : 'mean', 'start_lng' : 'mean'})
cs = cs.dropna()
cs = cs.reset_index()
cs['length_mean'] = cs['ride_length','mean']
cs['length_count'] = cs['ride_length','count']
cs['distance_mean'] = cs['distance']
cs = cs.drop(['ride_length','distance'], axis=1)
cs = cs.sort_values(by=['length_count','length_mean'], ascending=False)
cs ['s_lat'] = cs['start_lat','mean']
cs ['s_lng'] = cs['start_lng','mean']
cs = cs.drop(['start_lat','start_lng'], axis=1)
cs.nlargest(10, 'length_count')
```

```python
# create a member list for tableau data visualization
ms = member.groupby(['start_station_name']).agg(
    {"ride_length" : ["count","mean"],'distance':'mean','start_lat' : 'mean', 'start_lng' : 'mean'})
ms = ms.dropna()
ms = ms.reset_index()
ms['length_mean'] = ms['ride_length','mean']
ms['length_count'] = ms['ride_length','count']
ms['distance_mean'] = ms['distance']
ms = ms.drop(['ride_length','distance'], axis=1)
ms = ms.sort_values(by=['length_count','length_mean'], ascending=False)
ms ['s_lat'] = ms['start_lat','mean']
ms ['s_lng'] = ms['start_lng','mean']
ms = ms.drop(['start_lat','start_lng'], axis=1)
ms.nlargest(10, 'length_count')
```
